## lmplot() 

```python
sns.lmplot(x="KİŞİ BAŞINA",y="BÜYÜME",data=e)
```

```python
sns.lmplot(x="BÜYÜME",y="KİŞİ BAŞINA",data=e,hue="KITA")
```

```python
# col = "KITA" , her ülke için ayrı grafik oluşturur
sns.lmplot(x="BÜYÜME",y="KİŞİ BAŞINA",data=e,col="KITA")
```
