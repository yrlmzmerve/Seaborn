# Stripplot() ve Swarmplot()

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

import seaborn as sns
import warnings # uyarıları kaldırmak için

%matplotlib inline
```

```python
n=pd.read_csv("dünyam.csv")
n.dropna(how="any",inplace=True)
n.head(0)
```

```python
e=pd.read_csv("50ülke.csv")
e.pop("2023")
e.pop("SIRALAMA")
e.pop("SIRALAMA.2")
e.head(0)
```

```python
ev = pd.read_csv("evli.csv")
```

## Stripplot() 

```python
# Herveri bir nokta
# jitter = True : Dağılım halinde olur 

sns.stripplot(x="KITA",y="BÜYÜME",data=e,color="red",jitter=True)
```

## Swarmplot()

```python
sns.violinplot(x="KITA",y="BÜYÜME",data=e)
sns.swarmplot(x="KITA",y="BÜYÜME",data=e,color="red")
```
