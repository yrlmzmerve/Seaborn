## Factorplot()

```python
# kind=bar -> barplot haline gelir
sns.factorplot(x="KITA",y="BÜYÜME",data=e,kind="bar")
```

```python
# kind=violin -> violin haline gelir

sns.factorplot(x="KITA",y="BÜYÜME",data=e,kind="violin")
```

```python
# kind=point -> nokta halinde veri verir
sns.factorplot(x="KİŞİ BAŞINA",y="BÜYÜME",data=e,kind="point")
```

```python
t=pd.read_csv("titanic.csv")
t.head(0)
```

```python
sns.factorplot(x="Pclass",y="Survived",hue="Sex",size=4,data=t,kind="bar",palette="muted")
```

```python
sns.factorplot(x="Pclass",y="Survived",hue="Sex",size=6,data=t,kind="violin",palette="coolwarm")
```
