# Pairgrid() ve FacedGrid()

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

import seaborn as sns
import warnings # uyarıları kaldırmak için

%matplotlib inline
```

```python
n=pd.read_csv("dünyam.csv")
n.dropna(how="any",inplace=True)
```

## Pairgrid()

```python
# her birimin birbiri ile olan ilişkini grafikleştirir
m=sns.PairGrid(n)

# diag = orta , up=yukarı, low=aşağı
m.map_diag(sns.distplot)
m.map_upper(plt.scatter)
m.map_lower(sns.kdeplot)
```

```python
e=pd.read_csv("50ülke.csv")
e.pop("2023")
e.pop("SIRALAMA")
e.pop("SIRALAMA.2")
e.head(0)
```

```python
k=sns.PairGrid(e)
k.map_diag(sns.distplot)
k.map_upper(plt.scatter)
k.map_lower(sns.kdeplot)
```

## Facedgrid()

```python
s=pd.read_csv("evli.csv")
```

```python
ss=sns.FacetGrid(data=s,col="ay",row="evli")
ss.map(sns.distplot,"gelir")
```

```python
#histogram şeklinde yapma
ss=sns.FacetGrid(data=s,col="ay",row="evli")
ss.map(plt.hist,"gelir")
```
