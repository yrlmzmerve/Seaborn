# Pairplot() ve Rugplot()

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

import seaborn as sns
import warnings # uyarıları kaldırmak için

%matplotlib inline
```

```python
n=pd.read_csv("dünyam.csv")
n.dropna(how="any",inplace=True)
```

## pairplot()

```python
# n dosyasındaki tüm verileri birbiri ile kıyaslar

sns.pairplot(n,palette="navy")
```

```python
e=pd.read_csv("50ülke.csv")
e.pop("2023")
e.pop("SIRALAMA")
e.pop("SIRALAMA.2")
e.head()
```

```python
sns.pairplot(e, hue="KITA", palette="inferno")
```

## rugplot()

```python
sns.rugplot(e["BÜYÜME"], color="y" ,height=0.2)
sns.kdeplot(e["BÜYÜME"], color="blue")
```
