# Joinplot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

import seaborn as sns
import warnings # uyarıları kaldırmak için

%matplotlib inline
```

```python
n = pd.read_csv("dünyam.csv")
n.dropna(how="any", inplace=True)
```

```python
#2 veriyi birleştirme
sns.jointplot(n["Doğum oranı"],n["Ortalama yaş"], data=n)
```

```python
sns.jointplot(n["Yıllık değişim oranı"],n["Doğum oranı"],data=n)
```

## Yoğunlaşan yerleri data net görmeyi sağlar; kind:hex

```python
# kind ="hex"
sns.jointplot(n["Yıllık değişim oranı"],n["Doğum oranı"],data=n,kind="hex",color="orange")
```

```python
# kind="reg"
# xlim, ylim = hangi sayılar arasında olsun

sns.jointplot(n["Yıllık değişim oranı"],n["Doğum oranı"],data=n,kind="reg",xlim=(-2,6),ylim=(0,10),color="pink",size=8)
```

```python
#kind = kde
sns.jointplot(n["Yıllık değişim oranı"],n["Doğum oranı"],data=n,kind="kde",xlim=(-2,6),ylim=(0,10),color="b",size=8)
```
