## Dünyam.csv dosyasındaki verilerden grafik oluşturma

```python
n = pd.read_csv("dünyam.csv")
n.dropna(how="any", inplace=True)
```

```python
#bins = grafikteki mavi dikdötgenlerin kaç tane oldugu belirler
sns.distplot(n["Doğum oranı"], bins=20)
```

```python
# Grafiğin üstündeki mavi çizgiyi kaldırmak için : kde = false yapılır 
sns.distplot(n["Doğum oranı"], bins=20, kde=False, color="purple")
plt.tight_layout()
```

```python
# Grafiğin arkasındaki beyaz kısmı değiştirme : sns.set(style="dark")
# platte : renk
# font_scale : boyut 
sns.set(style="dark", palette="muted" , font_scale=1.5)

sns.distplot(n["Doğum oranı"], bins=20, kde=False, color="orange")

plt.tight_layout()
```

```python
# Grafiktek dikdörtgen gösterimi kapatmak için : hist=False
# Grafiğin altında her bölme için çizgi oluşturma : rug = True

sns.set(style="darkgrid", palette="pastel", font_scale=1.5)

sns.distplot(n["Yıllık değişim oranı"], rug=True, hist=False, color="red")

#plt.tight_layout()
```

```python
# Grafiğin altta kalan kısmını boyama : kde_kws={"shade":True}

sns.set(style="white", palette="Blues", font_scale=1.5)

sns.distplot(n["Ortalama yaş"],hist=False,color="green",kde_kws={"shade":True})

plt.tight_layout()
```

```python
sns.set(style="whitegrid", palette="Blues", font_scale=1.5)

sns.distplot(n["Nüfüs"],color="m")

plt.tight_layout()
```

## Yapılan grafikleri birleştirme

```python
sns.set(style="darkgrid", font_scale=1.5)

f, axes = plt.subplots(2,2,figsize=(15,10))

sns.distplot(n["Doğum oranı"], bins=20, kde=False, color="orange", ax=axes[0,0])

sns.distplot(n["Yıllık değişim oranı"], rug=True, hist=False, color="red", ax=axes[0,1])

sns.distplot(n["Nüfüs"],color="m", ax=axes[0,1])

sns.distplot(n["Ortalama yaş"],hist=False,color="green",kde_kws={"shade":True},)

```
